```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

# Sample dataset (features are just example numbers)
# Format: [weight, size, moisture], Class: 0 = dry waste, 1 = wet waste, 2 = recyclable
X = [
    [10, 2, 0], [15, 3, 1], [5, 1, 0], [30, 4, 1], [8, 1, 0],
    [12, 2, 0], [25, 5, 1], [7, 2, 0], [20, 4, 1], [9, 1, 0]
]
y = [2, 1, 0, 1, 0, 2, 1, 0, 1, 0]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Agent: Waste Classifier
agent = DecisionTreeClassifier()
agent.fit(X_train, y_train)

# Prediction
pred = agent.predict(X_test)

# Output: Classification Accuracy
accuracy = accuracy_score(y_test, pred)

print("Waste Segregation AI")
print("Classification Accuracy:", round(accuracy * 100, 2), "%")
```
