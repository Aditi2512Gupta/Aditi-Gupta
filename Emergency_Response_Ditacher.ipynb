```python
import numpy as np

np.random.seed(7)

num_vehicles = 5
num_incidents = 10

vehicles = np.random.randint(0, 50, size=(num_vehicles, 2))
incidents = np.random.randint(0, 50, size=(num_incidents, 2))

speed = 1.0

def distance(a, b):
    return np.sqrt(((a - b) ** 2).sum())

assigned_times = []

for i in range(num_incidents):
    incident = incidents[i]

    best_vehicle = None
    best_dist = float("inf")

    for v in range(num_vehicles):
        d = distance(incident, vehicles[v])

        if d < best_dist:
            best_dist = d
            best_vehicle = v

    response_time = best_dist / speed
    assigned_times.append(response_time)

    print(f"Incident {i+1} at {incident} assigned to Vehicle {best_vehicle+1} at {vehicles[best_vehicle]}")
    print(f"  → Distance = {best_dist:.2f}, Response Time = {response_time:.2f} seconds\n")

avg_time = sum(assigned_times) / len(assigned_times)

print("----------------------------------------------------")
print(f"Average Response Time: {avg_time:.2f} seconds")
print("----------------------------------------------------")
```
